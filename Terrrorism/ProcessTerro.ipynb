```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("gbtUS.csv",engine='python')
```

```python
date=[]
summary=[]
victim=[]
weapon=[]
propertyx=[]
newscover=[]
```

```python
for i in range(len(df)):
    date.append(str(df.iloc[i]["iyear"])+"-"+str(df.iloc[i]['imonth'])+"-"+str(df.iloc[i]['iday'])) 
    summary.append(str(df.iloc[i]['summary'])+":"+str(df.iloc[i]['addnotes']))
    victim.append(str(df.iloc[i]['target1'])+":"+str(df.iloc[i]["targtype1_txt"])+":"+str(df.iloc[i]['targsubtype1_txt']))
    weapon.append(str(df.iloc[i]['weaptype1_txt'])+":"+str(df.iloc[i]["weapdetail"]))
    propertyx.append(str(df.iloc[i]['propvalue'])+":"+str(df.iloc[i]['propcomment']))
    newscover.append(str(df.iloc[i]['scite1'])+":"+str(df.iloc[i]['scite2'])+":"+str(df.iloc[i]['scite3']))
```

```python
newdf=pd.DataFrame()
newdf['date']=date
newdf['city']=df['city']
newdf['state']=df['provstate']
newdf['long']=df['longitude']
newdf['latitude']=df['latitude']
newdf['summary']=summary
newdf['VicDescrition']=victim
newdf['VicNat']=df['natlty1_txt']
newdf['attacktype']=df['attacktype1_txt']
newdf['motive']=df['motive']
newdf['weapon']=weapon
newdf['corp']=df['corp1']
newdf['groupname']=df['gname']
newdf['injured']=df['nwound']
newdf['Property']=propertyx
newdf['NewsCover']=newscover
```
