# CrimeMap Plots

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("crime_full.csv")
```

```python
df['CrimeMap'].unique()
```

```python
df['CrimeMap'].value_counts().plot(kind='barh')
```

```python
data=df[df['CrimeMap']=='robbery']
data["CrimeType"].value_counts().plot(kind='barh')
percentage=len(data)/len(df)
plt.title('Robbery %f percent total'%(percentage*100))
plt.xlabel('Count')
plt.ylabel('CrimeType')
 
```
