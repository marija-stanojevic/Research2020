```python
import nltk 
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.stem.wordnet import WordNetLemmatizer
import gensim 
from gensim import corpora
import pandas as pd
import re 
import string


#from google.colab import drive
#drive.mount('/content/drive')
```

```python
def clean(doc):
    stop_free = " ".join([i for i in doc.lower().split() if i not in stop])
    punc_free = ''.join(ch for ch in stop_free if ch not in exclude)
    normalized = " ".join(lemma.lemmatize(word) for word in punc_free.split())
    return normalized
```

```python
abc=pd.read_csv("news_full.csv")
abc=abc.dropna(axis=0, how='any', thresh=None, subset=None, inplace=False)
```

```python
textdf=abc['Content']
finaltext=[]
for i in range(len(textdf)):
    if len(textdf[i])>1200:
        finaltext.append(i)
```

```python
abc=abc.iloc[finaltext]
textdf=abc['Content']
```

```python
topicArray=[]
count=6000
for text in textdf[6000:7000]:
    corpus = nltk.sent_tokenize(text)
    for i in range(len(corpus)):
          corpus [i] = corpus [i].lower()
          corpus [i] = re.sub(r'\W',' ',corpus [i])
          corpus [i] = re.sub(r'\s+',' ',corpus [i])
          corpus[i] = re.sub(r'[0-9]', '', corpus[i])
    corpus.remove(corpus[0])
    stop = set(stopwords.words('english'))
    exclude = set(string.punctuation)
    lemma = WordNetLemmatizer()

    corpus_clean = [clean(sent).split() for sent in corpus]
    dictionary = corpora.Dictionary(corpus_clean)
    doc_term_matrix = [dictionary.doc2bow(sent) for sent in corpus_clean]
    Lda = gensim.models.ldamodel.LdaModel
    ldamodel = Lda(doc_term_matrix, num_topics=30, id2word=dictionary, passes=1000)
    hold=ldamodel.print_topics(num_topics=30, num_words=1)
    Topic=[]
    for i in range(30):
        Topic.append(str(hold[i][1].split('*')[1][1:-1]))
    topicArray.append(Topic)
    print(count)
    count=count+1
    
    
```

```python
df1=abc[7000:9000]
```

```python
df1['LDA']=topicArray
```

```python
df1.to_csv("70009000.csv")
```
