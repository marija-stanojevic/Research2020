```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("database.csv")
```

```python
df.iloc[1]
```

```python
CrimeType=[]
for i in range(len(df)):
    CrimeType.append(df.iloc[i]['Crime Type']+":"+df.iloc[i]["Crime Solved"])
```

```python
Date=[]
for i in range(len(df)):
    Date.append(str(df.iloc[i]['Year'])+"-"+str(df.iloc[i]["Month"]))
```

```python
newdf=pd.DataFrame()
newdf['City']=df['City']
newdf["State"]=df['State']
newdf['Date']=Date
newdf['CrimeType']=CrimeType
newdf['VictimSex']=df['VictimSex']
newdf['VictimAge']=df['VictimAge']
newdf['VictimRace']=df['VictimRace']
newdf['PerpeSex']=df['Perpetrator Sex']
newdf['PerpeAge']=df['Perpetrator Age']
newdf["PerPeRace"]=df['Perpetrator Race']
newdf['Relationship']=df['Relationship']
newdf['']

```
